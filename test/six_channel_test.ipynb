```python
import numpy as np
import polyscope as ps
import potpourri3d as pp3d
```

```python
ps.init()
ps.set_ground_plane_mode("none")

# FPS for a full frame (one even and one odd frame)
ps.set_max_fps(30)

# Absolute clock, can call ps.show()
ps.set_render_even_odd_absolute_clock(True)

# May need to adjust this number in the range of (0, 100) to hit target FPS on a specific devide
# If the frame rate is lower than the target, decrease the target_sleep value
ps.set_target_sleep(98)

# Set always redraw to True (or else polyscope tries to optimize and won't draw even-odd correctly)
ps.set_always_redraw(True)

# Some debug options; also available in the Even-Odd gui panel
ps.set_draw_even_frame_first(True)
ps.set_black_out_even_frames(False)
ps.set_black_out_even_frames(False)

# Enable the Even-Odd gui panel
ps.set_build_even_odd_gui_panel(True)
```

```python
# Load a surface mesh
filename = "teapot.obj"
verts, faces = pp3d.read_mesh(filename)
ps_mesh = ps.register_surface_mesh("teapot", verts, faces, enabled=True)
ps_mesh.set_material("flat")

# Set six-channel color values
num_faces = ps_mesh.n_faces()

face_colors_even = np.ones((num_faces, 3)) * np.array([1.0, 0, 0.5])
face_colors_odd = np.ones((num_faces, 3)) * np.array([0.0, 0.5, 1.0])

ps_mesh.add_six_channel_color_quantity("face six channel", face_colors_even, face_colors_odd, 
                                       defined_on='faces', enabled=True)
```
